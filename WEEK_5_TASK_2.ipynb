```python
#import required libraries
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
page_no = 1
base_url = 'https://books.toscrape.com/'
url = base_url.format(page_no)

response = requests.get(url)
response
```

```python
print(response.text,'html.parser')
```

```python
soup = bs(response.text)
soup.find('title').text.strip()
```

```python
#lets try for ratings!
books_ratings = soup.find_all('p', class_='star-rating')
stars_display = {
    'One' : '1',
    'Two' : '2',
    'Three' : '3',
    'Four' : '4',
    'Five' : '5'
}
```

```python
len(books_ratings)
```

```python
for rating in books_ratings:
  star_text = rating['class'][1]
  star_value = stars_display.get(star_text)
  print(star_value)
```

```python
#now lets do this for links
#books_links = soup.find_all('a', href=re.compile(r'catalogue/.+\.html'))
#extracting urls from links
#links = [url + link['href'] for link in books_links if '/catalogue/category' not in link['href']]
#books_links = soup.find_all('a', href=True)
#links = [url + link['href'] for link in books_links if 'title' in link['href']]

#Let's try scrapping the link from image. Each image of the book has a link hence, will extract links for each book using the image!
image = soup.find_all('div', class_='image_container')
links = [url + container.find('a')['href'] for container in image]
```

```python
for link in links:
  print(link)
```
