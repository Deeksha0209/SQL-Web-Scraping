```python
#Lets get an overview of the drivers_license table wrt the interview transcripts
license = pd.read_sql_query('''
SELECT *
FROM drivers_license
WHERE plate_number LIKE '%H42W%' AND gender = 'male'
''',connection)
license
```

```python
#As displayed, both the ids are present in witness no 1 & witness no 2 qquery outputs! In order to narrow down our murderer, we need to use JOIN.
#Joining the tables wrt the filters used in where clause
pd.read_sql_query('''
SELECT *
FROM person as p
INNER JOIN drivers_license as dl on dl.id = p.license_id
INNER JOIN get_fit_now_member as gfnm on p.id = gfnm.person_id
INNER JOIN get_fit_now_check_in as gfnc on gfnc.membership_id = gfnm.id
WHERE plate_number LIKE '%H42W%' AND gender = 'male' AND membership_status = 'gold' AND check_in_date = 20180109
''',connection)

#Voila! We get our answer!! Jeremy Bowers!
```

```python
#Too many entries in the facebook table, however, Jeremy mentioned something about the income. Checking the income table
pd.read_sql_query('''
SELECT *
FROM income
''',connection)
```

```python
#income table didn't really provide much answers. In order to narrow down our suspects, we will use JOIN on drivers_license and facebook_event-checin wrt the filters provided.
pd.read_sql_query('''
SELECT *
FROM person as p
INNER JOIN drivers_license as dl on dl.id = p.license_id
INNER JOIN facebook_event_checkin as fb on p.id = fb.person_id
WHERE gender = 'female' AND hair_color = 'red' AND car_make = 'Tesla' AND car_model = 'Model S' AND event_name = 'SQL Symphony Concert' AND date LIKE '%201712%' AND height BETWEEN 65 AND 67
''',connection)

#Miranda Priestly it is! Murder mystery solved.
```
